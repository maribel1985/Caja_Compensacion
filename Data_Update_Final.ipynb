```python
# ==============================
# 4. CURSO 
# ==============================
# -------------------------
# CONEXIÓN
# -------------------------
conexion = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="Root8",
    database="caja_de_compensacion"
)

cursor = conexion.cursor()

print("Conexión exitosa")

import random
import itertools

cursos_ids = []

niveles = ["Básico", "Intermedio", "Avanzado", "Diplomado"]

tipos_curso = {
    "Educación": [
        "Fundamentos de Pedagogía",
        "Didáctica General",
        "Evaluación del Aprendizaje",
        "Diseño Curricular",
        "Planeación Educativa",
        "Estrategias de Enseñanza",
        "Educación Inclusiva",
        "Psicología del Aprendizaje",
        "Gestión Educativa",
        "Tecnología Aplicada a la Educación",
        "Metodologías Activas",
        "Investigación Educativa",
        "Desarrollo de Competencias Docentes",
        "Innovación Educativa",
        "Tutoría y Acompañamiento Escolar"
    ],
    "Capacitación laboral": [
        "Gestión de Proyectos",
        "Marketing Digital",
        "Atención al Cliente",
        "Excel Avanzado",
        "Contabilidad para Empresas",
        "Ventas Estratégicas",
        "Logística Empresarial",
        "Seguridad Industrial"
    ],
    "Bienestar": [
        "Mindfulness y Meditación",
        "Nutrición Saludable",
        "Yoga para Principiantes",
        "Manejo del Estrés",
        "Salud Mental",
        "Hábitos Saludables",
        "Entrenamiento Funcional",
        "Inteligencia Emocional"
    ],
    "Desarrollo personal": [
        "Liderazgo Personal",
        "Comunicación Asertiva",
        "Gestión del Tiempo",
        "Oratoria y Presentaciones",
        "Productividad Personal",
        "Toma de Decisiones",
        "Pensamiento Crítico",
        "Trabajo en Equipo"
    ]
}

# ----------------------------------------
# Funciones inteligentes
# ----------------------------------------

def generar_costo(nivel):
    if nivel == "Básico":
        return round(random.uniform(200000, 400000), 2)
    elif nivel == "Intermedio":
        return round(random.uniform(400000, 700000), 2)
    elif nivel == "Avanzado":
        return round(random.uniform(700000, 1000000), 2)
    else:  # Diplomado
        return round(random.uniform(1000000, 1500000), 2)

def generar_duracion(tipo):
    if tipo == "Educación":
        return random.randint(60, 120)
    elif tipo == "Capacitación laboral":
        return random.randint(40, 100)
    elif tipo == "Bienestar":
        return random.randint(30, 80)
    else:  # Desarrollo personal
        return random.randint(30, 90)

def generar_descripcion(nombre_base, nivel, tipo):
    return (
        f"El curso '{nombre_base}' en nivel {nivel} está diseñado dentro del área de {tipo}. "
        f"Brinda herramientas prácticas y teóricas para fortalecer competencias profesionales "
        f"mediante metodologías actualizadas y enfoque aplicado."
    )

# ----------------------------------------
# Generar combinaciones
# ----------------------------------------

todas_combinaciones = []

for tipo, cursos in tipos_curso.items():
    for nombre_base, nivel in itertools.product(cursos, niveles):
        nombre_completo = f"{nombre_base} - {nivel}"
        todas_combinaciones.append((nombre_completo, nombre_base, nivel, tipo))

random.shuffle(todas_combinaciones)

# ----------------------------------------
# Obtener existentes
# ----------------------------------------

cursor.execute("SELECT Nombre_Curso FROM curso")
existentes = {row[0] for row in cursor.fetchall()}

# ----------------------------------------
# Filtrar disponibles
# ----------------------------------------

disponibles = [
    combo for combo in todas_combinaciones
    if combo[0] not in existentes
]

cantidad_a_insertar = 40

if len(disponibles) < cantidad_a_insertar:
    raise ValueError("No hay suficientes cursos disponibles para insertar.")

seleccionados = disponibles[:cantidad_a_insertar]

# ----------------------------------------
# Insertar (sin ID si es AUTO_INCREMENT)
# ----------------------------------------

for nombre_completo, nombre_base, nivel, tipo in seleccionados:

    sql = """
    INSERT INTO curso (Nombre_Curso, Tipo_Curso, Descripcion, Duracion, Costo)
    VALUES (%s, %s, %s, %s, %s)
    """

    cursor.execute(sql, (
        nombre_completo,
        tipo,
        generar_descripcion(nombre_base, nivel, tipo),
        generar_duracion(tipo),
        generar_costo(nivel)
    ))

    cursos_ids.append(cursor.lastrowid)
    
# Guardar cambios
conexion.commit()

print("Datos insertados correctamente")

# Cerrar conexión
cursor.close()
conexion.close()
print('Conexión cerrada')
```

```python
# ==============================
# 5. HORARIO_CURSO (40)
# ==============================
import mysql.connector
from datetime import timedelta, datetime
import random

conexion = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="Root8",
    database="caja_de_compensacion"
)

cursor = conexion.cursor()
print("Conexión exitosa")

sedes = [
    "Sede Centro",
    "Sede Norte",
    "Sede Sur",
    "Sede Occidente",
    "Sede Empresarial"
]

hoy = datetime.today().date()

cursor.execute("SELECT ID_Curso, Duracion FROM curso")
cursos = cursor.fetchall()

total_cursos = len(cursos)

for i, (curso_id, duracion_original) in enumerate(cursos):

    modalidad = random.choice(["Presencial", "Virtual"])
    ubicacion = random.choice(sedes) if modalidad == "Presencial" else None

    jornada = random.choice(["Nocturno", "Sabado"])

    if jornada == "Nocturno":
        horas_por_dia = 3
        dias_clase_numeros = [2, 4]
        dias_clase_str = "Miercoles,Viernes"
        hora_inicio = random.choice(["18:00:00", "19:00:00"])
        hora_fin = "21:00:00" if hora_inicio == "18:00:00" else "22:00:00"
    else:
        horas_por_dia = 5
        dias_clase_numeros = [5]
        dias_clase_str = "Sabado"
        hora_inicio = random.choice(["08:00:00", "13:00:00"])
        hora_fin = "13:00:00" if hora_inicio == "08:00:00" else "18:00:00"

    # Ajustar duración al múltiplo exacto
    duracion_ajustada = (duracion_original // horas_por_dia) * horas_por_dia

    if duracion_ajustada == 0:
        duracion_ajustada = horas_por_dia

    sesiones = duracion_ajustada // horas_por_dia

    # 🔥 ACTUALIZAR tabla curso
    cursor.execute(
        "UPDATE curso SET Duracion = %s WHERE ID_Curso = %s",
        (duracion_ajustada, curso_id)
    )

    porcentaje = i / total_cursos

    if porcentaje < 0.70:
        fecha_base = hoy - timedelta(days=random.randint(60, 180))
    elif porcentaje < 0.90:
        fecha_base = hoy - timedelta(days=random.randint(5, 30))
    else:
        fecha_base = hoy + timedelta(days=random.randint(5, 60))

    fechas_clase = []
    fecha_cursor = fecha_base

    while len(fechas_clase) < sesiones:
        if fecha_cursor.weekday() in dias_clase_numeros:
            fechas_clase.append(fecha_cursor)
        fecha_cursor += timedelta(days=1)

    fecha_inicio = fechas_clase[0]
    fecha_fin = fechas_clase[-1]

    sql = """
    INSERT INTO horario_curso 
    (ID_Curso, Fecha_Inicio, Fecha_Fin, Hora_Inicio, Hora_Fin, Modalidad, Ubicacion, Dias_Clase)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """

    cursor.execute(sql, (
        curso_id,
        fecha_inicio,
        fecha_fin,
        hora_inicio,
        hora_fin,
        modalidad,
        ubicacion,
        dias_clase_str
    ))

conexion.commit()
print("Horarios generados y duración sincronizada con tabla curso")

cursor.close()
conexion.close()
print("Conexión cerrada")
```

```python
# ==============================
# 5.1. HORARIO_CURSO 
# ==============================
import mysql.connector
from datetime import timedelta, datetime

conexion = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="Root8",
    database="caja_de_compensacion"
)

cursor = conexion.cursor()

# Traer info completa
query = """
SELECT h.ID_Curso, h.Fecha_Inicio, h.Fecha_Fin,
       h.Hora_Inicio, h.Hora_Fin, h.Dias_Clase,
       c.Duracion
FROM horario_curso h
JOIN curso c ON h.ID_Curso = c.ID_Curso
"""

cursor.execute(query)
datos = cursor.fetchall()

errores = 0

for fila in datos:
    curso_id, fecha_inicio, fecha_fin, hora_inicio, hora_fin, dias_clase_str, duracion_real = fila

    # Convertir horas a datetime
    hora_inicio_dt = datetime.strptime(str(hora_inicio), "%H:%M:%S")
    hora_fin_dt = datetime.strptime(str(hora_fin), "%H:%M:%S")

    horas_por_dia = (hora_fin_dt - hora_inicio_dt).seconds / 3600

    # Mapear días
    mapa_dias = {
        "Lunes": 0,
        "Martes": 1,
        "Miercoles": 2,
        "Jueves": 3,
        "Viernes": 4,
        "Sabado": 5
    }

    dias_validos = [mapa_dias[d] for d in dias_clase_str.split(",")]

    # Contar sesiones reales
    sesiones = 0
    fecha_cursor = fecha_inicio

    while fecha_cursor <= fecha_fin:
        if fecha_cursor.weekday() in dias_validos:
            sesiones += 1
        fecha_cursor += timedelta(days=1)

    horas_calculadas = sesiones * horas_por_dia

    if horas_calculadas != duracion_real:
        print(f"Inconsistencia en curso {curso_id}")
        print(f"Duración esperada: {duracion_real}")
        print(f"Duración calculada: {horas_calculadas}")
        print("-----------------------------")
        errores += 1

if errores == 0:
    print("✅ Todas las duraciones coinciden perfectamente.")
else:
    print(f"⚠ Se encontraron {errores} inconsistencias.")

cursor.close()
conexion.close()
```

```python
# ==============================
# 6.1. UPDATE TABLA CLIENTE
# Genero, Edad, Estrato, Nivel_Educativo
# ==============================

import mysql.connector
import random

conexion = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="Root8",
    database="caja_de_compensacion"
)

cursor = conexion.cursor(dictionary=True)
print("Conexión exitosa")

# -----------------------------------------------
# 1. LISTAS DE NOMBRES PARA INFERIR GÉNERO
# -----------------------------------------------

nombres_femeninos = {
    "María", "Ana", "Laura", "Claudia", "Sandra", "Patricia", "Gloria",
    "Marcela", "Adriana", "Diana", "Paola", "Luisa", "Isabel", "Valentina",
    "Natalia", "Carolina", "Alejandra", "Daniela", "Camila", "Sofía",
    "Juliana", "Andrea", "Mónica", "Liliana", "Ángela", "Lorena", "Martha",
    "Catalina", "Sara", "Yolanda", "Blanca", "Rosa", "Elena", "Paula",
    "Fernanda", "Johanna", "Viviana", "Nathalia", "Mariana", "Manuela",
    "Pilar", "Esperanza", "Fabiola", "Xiomara", "Luz", "Nubia", "Olga",
    "Beatriz", "Amparo", "Cecilia", "Nora", "Lina", "Tatiana", "Sonia"
}

nombres_masculinos = {
    "Carlos", "Andrés", "Luis", "Jorge", "Juan", "Miguel", "Sergio",
    "Diego", "Felipe", "Alejandro", "Ricardo", "Jhon", "David", "Oscar",
    "Gabriel", "Sebastián", "Camilo", "Daniel", "Mauricio", "Álvaro",
    "Hernando", "Fernando", "Édgar", "Iván", "Nelson", "Gustavo", "Jairo",
    "Henry", "Francisco", "Roberto", "Rodrigo", "Ramón", "William",
    "Cristian", "Jhonatan", "Esteban", "Nicolás", "Fabio", "Javier",
    "Héctor", "Pedro", "Manuel", "Raúl", "Edwin", "Víctor", "Germán",
    "Ernesto", "Armando", "Rubén", "Bernardo", "Alfredo", "Ignacio"
}

def inferir_genero(nombre_completo):
    """Toma el primer nombre del cliente e infiere el género."""
    primer_nombre = nombre_completo.strip().split()[0]
    if primer_nombre in nombres_femeninos:
        return "F"
    elif primer_nombre in nombres_masculinos:
        return "M"
    else:
        # Si no está en ninguna lista, asignar aleatoriamente con leve sesgo masculino
        return random.choice(["F", "F", "M", "M", "M"])

# -----------------------------------------------
# 2. DISTRIBUCIÓN DE ESTRATOS (contexto colombiano)
# Grupos: A (extrema pobreza), B (pobreza moderada),
#         C (vulnerable), D (no pobre, no vulnerable)
# Subgrupos: A1-A5, B1-B7, C1-C18, D1-D21
# Distribución aprox. Colombia urbana:
#   A: ~10%, B: ~15%, C: ~40%, D: ~35%
# -----------------------------------------------

def generar_estrato():
    prob = random.random()
    if prob < 0.10:
        grupo = "A"
        subgrupo = random.randint(1, 5)
    elif prob < 0.25:
        grupo = "B"
        subgrupo = random.randint(1, 7)
    elif prob < 0.65:
        grupo = "C"
        subgrupo = random.randint(1, 18)
    else:
        grupo = "D"
        subgrupo = random.randint(1, 21)
    return f"{grupo}{subgrupo}"

# -----------------------------------------------
# 3. NIVEL EDUCATIVO según estrato
# Grupo A → alta prob. básica/media
# Grupo B → media/técnica
# Grupo C → técnica/tecnológica/pregrado
# Grupo D → pregrado/posgrado
# -----------------------------------------------

niveles_por_grupo = {
    "A": {
        "opciones": ["Básica", "Media", "Técnica"],
        "pesos":    [0.50,    0.35,   0.15]
    },
    "B": {
        "opciones": ["Básica", "Media", "Técnica", "Tecnológica"],
        "pesos":    [0.20,    0.40,   0.25,       0.15]
    },
    "C": {
        "opciones": ["Media", "Técnica", "Tecnológica", "Pregrado"],
        "pesos":    [0.15,   0.25,      0.30,          0.30]
    },
    "D": {
        "opciones": ["Tecnológica", "Pregrado", "Posgrado"],
        "pesos":    [0.10,         0.55,       0.35]
    }
}

def generar_nivel_educativo(estrato_codigo):
    grupo = estrato_codigo[0]  # Toma la letra: A, B, C o D
    config = niveles_por_grupo[grupo]
    return random.choices(config["opciones"], weights=config["pesos"], k=1)[0]

# -----------------------------------------------
# 4. OBTENER TODOS LOS CLIENTES
# -----------------------------------------------

cursor.execute("SELECT ID_Cliente, Nombre_Cliente FROM cliente")
clientes = cursor.fetchall()

print(f"Total clientes a actualizar: {len(clientes)}")

# -----------------------------------------------
# 5. ACTUALIZAR CADA CLIENTE
# -----------------------------------------------

actualizados = 0

for cliente in clientes:
    id_cliente = cliente["ID_Cliente"]
    nombre = cliente["Nombre_Cliente"]

    genero = inferir_genero(nombre)

    # Edad según género
    if genero == "F":
        edad = random.randint(18, 55)
    else:
        edad = random.randint(18, 62)

    estrato = generar_estrato()
    nivel_educativo = generar_nivel_educativo(estrato)

    cursor.execute("""
        UPDATE cliente
        SET Genero = %s,
            Edad = %s,
            Estrato = %s,
            Nivel_Educativo = %s
        WHERE ID_Cliente = %s
    """, (genero, edad, estrato, nivel_educativo, id_cliente))

    actualizados += 1

conexion.commit()
print(f"✅ {actualizados} clientes actualizados correctamente.")

# -----------------------------------------------
# 6. VALIDACIÓN RÁPIDA
# -----------------------------------------------

cursor.execute("""
    SELECT
        Genero,
        COUNT(*) AS Total,
        ROUND(AVG(Edad), 1) AS Edad_Promedio,
        LEFT(Estrato, 1) AS Grupo
    FROM cliente
    GROUP BY Genero, LEFT(Estrato, 1)
    ORDER BY Genero, Grupo
""")

print("\n📊 Distribución resultante:")
print(f"{'Género':<10} {'Grupo':<8} {'Total':<10} {'Edad Prom.'}")
for fila in cursor.fetchall():
    print(f"{fila['Genero']:<10} {fila['Grupo']:<8} {fila['Total']:<10} {fila['Edad_Promedio']}")

cursor.close()
conexion.close()
print("\n🔒 Conexión cerrada.")
```

```python
# ==============================
# 7. MATRICULA
# ==============================
# -------------------------
# CONEXIÓN
# -------------------------
conexion = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="Root8",
    database="caja_de_compensacion"
)

cursor = conexion.cursor()

print("Conexión exitosa")

import random
from faker import Faker
from datetime import timedelta

fake = Faker("es_CO")

# =========================================
# 1️⃣ OBTENER DATOS DE LA BD
# =========================================

# Clientes
cursor.execute("SELECT ID_Cliente, Direccion_Cliente FROM Cliente")
clientes = cursor.fetchall()

# Cursos + Modalidad + Fecha_inicio
cursor.execute("""
    SELECT 
        c.ID_Curso,
        c.Tipo_Curso,
        h.Modalidad,
        h.Fecha_Inicio
    FROM Curso c
    JOIN Horario_Curso h ON c.ID_Curso = h.ID_Curso
""")
cursos = cursor.fetchall()

# Formadores
cursor.execute("SELECT ID_Formador, Especializacion FROM Formador")
formadores = cursor.fetchall()


# =========================================
# 2️⃣ OPTIMIZAR BÚSQUEDAS CON DICCIONARIOS
# =========================================

clientes_dict = {c[0]: c[1] for c in clientes}
formadores_por_tipo = {}

for id_formador, especializacion in formadores:
    if especializacion not in formadores_por_tipo:
        formadores_por_tipo[especializacion] = []
    formadores_por_tipo[especializacion].append(id_formador)


# =========================================
# 3️⃣ PREPARAR 624 MATRÍCULAS
# =========================================

clientes_ids = list(clientes_dict.keys())

clientes_para_doble = random.sample(clientes_ids, 124)

lista_matriculas = clientes_ids.copy()

for cliente in clientes_para_doble:
    lista_matriculas.append(cliente)

print("Total matrículas a generar:", len(lista_matriculas))


# =========================================
# 4️⃣ FUNCIÓN VALIDAR MEDELLÍN
# =========================================

def es_medellin(direccion):
    zonas = [
        "medellin", "bello", "itagui",
        "envigado", "sabaneta", "la estrella",
        "copacabana", "girardota"
    ]
    direccion = direccion.lower()
    return any(z in direccion for z in zonas)


# =========================================
# 5️⃣ INSERTAR MATRÍCULAS
# =========================================

matriculas_ids = []
cursos_por_cliente = {}

for id_cliente in lista_matriculas:

    direccion = clientes_dict[id_cliente]
    cliente_medellin = es_medellin(direccion)

    if id_cliente not in cursos_por_cliente:
        cursos_por_cliente[id_cliente] = []

    if cliente_medellin:
        cursos_validos = cursos
    else:
        cursos_validos = [c for c in cursos if c[2] == "Virtual"]

    cursos_validos = [
        c for c in cursos_validos
        if c[0] not in cursos_por_cliente[id_cliente]
    ]

    if not cursos_validos:
        continue

    curso_elegido = random.choice(cursos_validos)

    id_curso     = curso_elegido[0]
    tipo_curso   = curso_elegido[1]
    modalidad    = curso_elegido[2]
    fecha_inicio = curso_elegido[3]

    if tipo_curso not in formadores_por_tipo:
        continue

    id_formador = random.choice(formadores_por_tipo[tipo_curso])

    fecha_limite   = fecha_inicio - timedelta(days=1)
    fecha_matricula = fake.date_between(
        start_date='-1y',
        end_date=fecha_limite
    )

    # ← Todo esto también va dentro del for, con indentación correcta
    cursor.execute("""
        SELECT c.Costo, cl.Tipo_Cliente, cl.Estrato
        FROM curso c, cliente cl
        WHERE c.ID_Curso = %s AND cl.ID_Cliente = %s
    """, (id_curso, id_cliente))

    datos        = cursor.fetchone()
    costo        = float(datos[0])
    tipo_cliente = datos[1]
    estrato      = datos[2]

    descuentos = {"A": 0.70, "B": 0.50, "C": 0.30, "D": 0.10}
    grupo = estrato[0] if estrato else None

    if tipo_cliente == "Corporativo" and grupo in descuentos:
        descuento = descuentos[grupo]
    else:
        descuento = 0.0

    monto_final = round(costo * (1 - descuento), 2)

    sql = """
    INSERT INTO Matricula 
    (ID_Cliente, ID_Curso, ID_Formador, Fecha_Matricula,
     Descuento_Aplicado, Monto_Final)
    VALUES (%s, %s, %s, %s, %s, %s)
    """

    cursor.execute(sql, (
        id_cliente,
        id_curso,
        id_formador,
        fecha_matricula,
        descuento,
        monto_final
    ))

    matriculas_ids.append(cursor.lastrowid)
    cursos_por_cliente[id_cliente].append(id_curso)

# Guardar cambios
conexion.commit()

print("✅ Matrículas insertadas correctamente:", len(matriculas_ids))

# Cerrar conexión
cursor.close()
conexion.close()
print('Conexión cerrada')
```

```python
# ==============================
# 8. PAGO
# ==============================
import mysql.connector
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import random

print("🔎 Regenerando tabla pago correctamente (lógica realista)...")

try:
    conexion = mysql.connector.connect(
        host="localhost",
        port=3306,
        user="root",
        password="Root8",
        database="caja_de_compensacion"
    )

    cursor = conexion.cursor(dictionary=True)
    hoy = datetime.today().date()

    # Limpiar tabla completamente
    cursor.execute("DELETE FROM pago")
    conexion.commit()
    print("🗑 Tabla pago limpiada.")

    query = """
    SELECT 
        m.ID_Matricula,
        m.Monto_Final AS Costo,
        h.Fecha_Inicio,
        h.Fecha_Fin
    FROM matricula m
    JOIN curso c ON m.ID_Curso = c.ID_Curso
    JOIN horario_curso h ON c.ID_Curso = h.ID_Curso
    """

    cursor.execute(query)
    matriculas = cursor.fetchall()

    for m in matriculas:

        id_matricula = m["ID_Matricula"]
        costo        = float(m["Costo"])
        fecha_inicio = m["Fecha_Inicio"]
        fecha_fin    = m["Fecha_Fin"]

        fecha_primer_vencimiento = fecha_inicio - relativedelta(months=1)

        metodo = random.choice([
            "Efectivo",
            "Tarjeta Crédito",
            "Tarjeta Débito",
            "PSE"
        ])

        curso_no_iniciado = fecha_inicio > hoy
        curso_finalizado  = fecha_fin < hoy

        # ==============================
        # CASO 1: CURSO NO HA INICIADO
        # ==============================
        if curso_no_iniciado:

            insert = """
            INSERT INTO pago
            (ID_Matricula, Monto, Metodo, Estado, Fecha_Pago,
             Fecha_Maxima_Pago, Canal, Numero_Cuota, Total_Cuotas)
            VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s)
            """

            cursor.execute(insert, (
                id_matricula,
                costo,
                metodo,
                "Pendiente",
                None,
                fecha_primer_vencimiento,
                None,
                1,
                1
            ))

            continue

        # ==============================
        # CASO 2: CURSO YA INICIÓ
        # ==============================

        total_cuotas = random.choice([1, 3, 5])
        valor_base   = round(costo / total_cuotas, 0)
        ultima_cuota = costo - (valor_base * (total_cuotas - 1))

        generar_mas_cuotas = True

        for n in range(1, total_cuotas + 1):

            if not generar_mas_cuotas:
                break

            monto             = ultima_cuota if n == total_cuotas else valor_base
            fecha_maxima_pago = fecha_primer_vencimiento + relativedelta(months=n-1)

            estado     = "Pendiente"
            fecha_pago = None
            canal      = None

            rango_dias = (fecha_maxima_pago - fecha_primer_vencimiento).days
            if rango_dias > 0:
                fecha_generada = fecha_primer_vencimiento + timedelta(
                    days=random.randint(0, rango_dias)
                )
            else:
                fecha_generada = fecha_maxima_pago

            # Curso finalizado
            if curso_finalizado:

                prob = random.random()

                if prob < 0.15:
                    estado             = "En Mora"
                    generar_mas_cuotas = False

                elif prob < 0.30:
                    estado             = "Cancelado"
                    fecha_pago         = fecha_generada
                    generar_mas_cuotas = False

                else:
                    estado     = "Completado"
                    fecha_pago = fecha_generada

            # Curso en curso
            else:

                prob = random.random()

                if prob < 0.6:
                    estado     = "Completado"
                    fecha_pago = fecha_generada
                else:
                    estado             = "Pendiente"
                    generar_mas_cuotas = False

            # Canal solo si efectivo y completado
            if metodo == "Efectivo" and estado == "Completado":
                canal = random.choice([
                    "Corresponsal Bancario",
                    "Sucursal Bancaria",
                    "SuperGIROS",
                    "Efecty"
                ])

            if estado == "En Mora":
                fecha_pago = None
                canal      = None

            insert = """
            INSERT INTO pago
            (ID_Matricula, Monto, Metodo, Estado, Fecha_Pago,
             Fecha_Maxima_Pago, Canal, Numero_Cuota, Total_Cuotas)
            VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s)
            """

            cursor.execute(insert, (
                id_matricula,
                monto,
                metodo,
                estado,
                fecha_pago,
                fecha_maxima_pago,
                canal,
                n,
                total_cuotas
            ))

    conexion.commit()
    print("✅ Pagos regenerados correctamente con lógica realista.")

except Exception as e:
    print("❌ Error:", e)

finally:
    cursor.close()
    conexion.close()
    print("🔒 Conexión cerrada.")
```

```python
# ==============================
# 8.1. PAGO
# ==============================
import mysql.connector

try:
    conexion = mysql.connector.connect(
        host="localhost",
        port=3306,
        user="root",
        password="Root8",
        database="caja_de_compensacion"
    )

    cursor = conexion.cursor(dictionary=True)
    print("🔎 Iniciando auditoría financiera...\n")

    inconsistencias = 0

    # 1️⃣ Efectivo en mora con canal
    cursor.execute("""
        SELECT * FROM pago
        WHERE Metodo = 'Efectivo'
        AND Estado = 'En Mora'
        AND Canal IS NOT NULL
    """)
    r = cursor.fetchall()
    if r:
        print("❌ Efectivo en mora con canal detectado")
        inconsistencias += len(r)

    # 2️⃣ Efectivo con estado cancelado
    cursor.execute("""
        SELECT * FROM pago
        WHERE Metodo = 'Efectivo'
        AND Estado = 'Cancelado'
    """)
    r = cursor.fetchall()
    if r:
        print("❌ Pago en efectivo con estado cancelado detectado")
        inconsistencias += len(r)

    # 3️⃣ En mora con fecha de pago
    cursor.execute("""
        SELECT * FROM pago
        WHERE Estado = 'En Mora'
        AND Fecha_Pago IS NOT NULL
    """)
    r = cursor.fetchall()
    if r:
        print("❌ Pago en mora con fecha registrada")
        inconsistencias += len(r)

    # 4️⃣ Sobrepago
    cursor.execute("""
        SELECT 
            m.ID_Matricula,
            c.Costo,
            SUM(CASE WHEN p.Estado='Completado' THEN p.Monto ELSE 0 END) AS Total_Pagado
        FROM matricula m
        JOIN curso c ON m.ID_Curso=c.ID_Curso
        JOIN pago p ON m.ID_Matricula=p.ID_Matricula
        GROUP BY m.ID_Matricula, c.Costo
        HAVING Total_Pagado > c.Costo
    """)
    r = cursor.fetchall()
    if r:
        print("❌ Matrículas con sobrepago detectadas")
        inconsistencias += len(r)

    # 5️⃣ Curso finalizado sin todas las cuotas completadas (sin mora ni cancelado)
    cursor.execute("""
        SELECT 
            m.ID_Matricula,
            COUNT(p.ID_Pago) AS Total_Registros,
            SUM(CASE WHEN p.Estado='Completado' THEN 1 ELSE 0 END) AS Completados
        FROM matricula m
        JOIN curso c ON m.ID_Curso=c.ID_Curso
        JOIN horario_curso h ON c.ID_Curso=h.ID_Curso
        JOIN pago p ON m.ID_Matricula=p.ID_Matricula
        WHERE h.Fecha_Fin < CURDATE()
        GROUP BY m.ID_Matricula
        HAVING Completados < Total_Registros
    """)
    r = cursor.fetchall()
    if r:
        print("⚠ Cursos finalizados con cuotas no completadas")
        inconsistencias += len(r)

    # 6️⃣ Curso finalizado con suma distinta al costo
    cursor.execute("""
        SELECT 
            m.ID_Matricula,
            c.Costo,
            SUM(CASE WHEN p.Estado='Completado' THEN p.Monto ELSE 0 END) AS Total_Pagado
        FROM matricula m
        JOIN curso c ON m.ID_Curso=c.ID_Curso
        JOIN horario_curso h ON c.ID_Curso=h.ID_Curso
        JOIN pago p ON m.ID_Matricula=p.ID_Matricula
        WHERE h.Fecha_Fin < CURDATE()
        GROUP BY m.ID_Matricula, c.Costo
        HAVING Total_Pagado <> c.Costo
    """)
    r = cursor.fetchall()
    if r:
        print("❌ Curso finalizado con total pagado distinto al costo")
        inconsistencias += len(r)

    # 7️⃣ Más de un método por matrícula
    cursor.execute("""
        SELECT ID_Matricula
        FROM pago
        GROUP BY ID_Matricula
        HAVING COUNT(DISTINCT Metodo) > 1
    """)
    r = cursor.fetchall()
    if r:
        print("❌ Matrículas con más de un método de pago")
        inconsistencias += len(r)

    print("\n==============================")
    if inconsistencias == 0:
        print("✅ Auditoría finalizada: SIN inconsistencias")
    else:
        print(f"⚠ Auditoría finalizada con {inconsistencias} inconsistencias detectadas")

except Exception as e:
    print("Error en auditoría:", e)

finally:
    cursor.close()
    conexion.close()
    print("Conexión cerrada")
```

```python
# ==============================
# 9. ASOCIACION CLIENTE EMPRESA
# ==============================
import mysql.connector
import random
from datetime import timedelta

print("🔎 Iniciando población de asociacion_cliente_empresa...")

try:
    conexion = mysql.connector.connect(
        host="localhost",
        port=3306,
        user="root",
        password="Root8",
        database="caja_de_compensacion"
    )

    cursor = conexion.cursor()

    cargos_por_nivel = {
        "Básica"      : ["Operario", "Auxiliar", "Asistente"],
        "Media"       : ["Operario", "Auxiliar", "Asistente"],
        "Técnica"     : ["Técnico", "Asistente", "Auxiliar"],
        "Tecnológica" : ["Tecnólogo", "Técnico", "Asistente", "Administrativo"],
        "Pregrado"    : ["Analista", "Coordinador", "Administrativo", "Consultor"],
        "Posgrado"    : ["Gerente", "Director", "Supervisor", "Consultor"]
    }
    cargos_genericos = ["Asistente", "Auxiliar", "Administrativo"]

    # Obtener clientes con empresa y su primera fecha de matrícula
    cursor.execute("""
        SELECT c.id_cliente, c.id_empresa, MIN(m.fecha_matricula) AS primera_matricula
        FROM cliente c
        JOIN matricula m ON c.id_cliente = m.id_cliente
        WHERE c.id_empresa IS NOT NULL
        GROUP BY c.id_cliente, c.id_empresa
    """)

    clientes = cursor.fetchall()
    asociaciones_insertadas = 0

    for cliente in clientes:

        id_cliente      = cliente[0]
        id_empresa      = cliente[1]
        fecha_matricula = cliente[2]

        # Fecha asociación mínimo 15 días antes de la primera matrícula
        fecha_asociacion = fecha_matricula - timedelta(days=random.randint(15, 90))

        estado = "Activo"

        # Obtener nivel educativo del cliente
        cursor.execute("""
            SELECT Nivel_Educativo
            FROM cliente
            WHERE ID_Cliente = %s
        """, (id_cliente,))

        resultado       = cursor.fetchone()
        nivel_educativo = resultado[0] if resultado else None

        cargos_validos = cargos_por_nivel.get(nivel_educativo, cargos_genericos)
        cargo          = random.choice(cargos_validos)

        cursor.execute("""
            INSERT INTO asociacion_cliente_empresa
            (ID_Cliente, ID_Empresa, Cargo, Estado, Fecha_Asociacion)
            VALUES (%s, %s, %s, %s, %s)
        """, (id_cliente, id_empresa, cargo, estado, fecha_asociacion))

        asociaciones_insertadas += 1

    conexion.commit()
    print(f"✅ Se insertaron {asociaciones_insertadas} asociaciones correctamente.")

    # Validación contra empleados_inscritos
    print("\n🔎 Validando cantidades por empresa...")

    cursor.execute("""
        SELECT ec.id_empresa, ec.empleados_inscritos,
               COUNT(a.ID_Cliente) AS total_asociados
        FROM empresa_cliente ec
        LEFT JOIN asociacion_cliente_empresa a
            ON ec.id_empresa = a.ID_Empresa
        GROUP BY ec.id_empresa, ec.empleados_inscritos
    """)

    inconsistencias = 0

    for fila in cursor.fetchall():
        id_empresa          = fila[0]
        empleados_inscritos = fila[1]
        total_asociados     = fila[2]

        if empleados_inscritos != total_asociados:
            print(f"❌ Empresa {id_empresa}: Esperados {empleados_inscritos}, Asociados {total_asociados}")
            inconsistencias += 1

    if inconsistencias == 0:
        print("🎉 Validación correcta. Todo coincide.")
    else:
        print(f"⚠ Se detectaron {inconsistencias} inconsistencias.")

except Exception as e:
    conexion.rollback()
    print("❌ Error detectado:", e)

finally:
    cursor.close()
    conexion.close()
    print("🔒 Conexión cerrada.")
```

```python
# ==============================
# 3.1. CALIFICACION PROMEDIO FORMADOR
# ==============================

import mysql.connector

try:
    conexion = mysql.connector.connect(
        host="localhost",
        port=3306,
        user="root",
        password="Root8",
        database="caja_de_compensacion"
    )

    cursor = conexion.cursor()
    print("Conexión exitosa")

    # Calcular promedio de calificaciones por formador
    # Ruta: Calificaciones_Al_Formador → Matricula → Formador
    cursor.execute("""
        SELECT
            m.ID_Formador,
            ROUND(AVG(cf.Calificacion), 2) AS Promedio
        FROM calificaciones_al_formador cf
        JOIN matricula m ON cf.ID_Matricula = m.ID_Matricula
        WHERE cf.Calificacion IS NOT NULL
        GROUP BY m.ID_Formador
    """)

    promedios = cursor.fetchall()
    print(f"Formadores con calificaciones para actualizar: {len(promedios)}")

    actualizados = 0

    for id_formador, promedio in promedios:
        cursor.execute("""
            UPDATE formador
            SET Calificacion = %s
            WHERE ID_Formador = %s
        """, (promedio, id_formador))
        actualizados += 1

    conexion.commit()
    print(f"✅ {actualizados} formadores actualizados con su calificación promedio.")

    # Validación rápida
    cursor.execute("""
        SELECT
            ID_Formador,
            Calificacion
        FROM formador
        WHERE Calificacion IS NOT NULL
        ORDER BY Calificacion DESC
        LIMIT 5
    """)
    print("\n📊 Top 5 formadores por calificación:")
    print(f"{'ID_Formador':<15} {'Calificación'}")
    print("-" * 30)
    for fila in cursor.fetchall():
        print(f"{fila[0]:<15} {fila[1]}")

except Exception as e:
    conexion.rollback()
    print("❌ Error:", e)

finally:
    cursor.close()
    conexion.close()
    print("🔒 Conexión cerrada.")
```

```python
# ======================================================
# 12. ASISTENCIA_ESTUDIANTE
# ======================================================

import mysql.connector
import random
from datetime import datetime

try:
    conexion = mysql.connector.connect(
        host="localhost",
        port=3306,
        user="root",
        password="Root8",
        database="caja_de_compensacion"
    )

    cursor = conexion.cursor(dictionary=True)
    hoy = datetime.today().date()
    print("Conexión exitosa")

    cursor.execute("""
        SELECT
            m.ID_Matricula,
            c.Duracion,
            h.Hora_Inicio,
            h.Hora_Fin,
            h.Fecha_Fin
        FROM matricula m
        JOIN curso c ON m.ID_Curso = c.ID_Curso
        JOIN horario_curso h ON c.ID_Curso = h.ID_Curso
    """)

    registros = cursor.fetchall()
    insertados = 0

    for r in registros:

        id_matricula     = r["ID_Matricula"]
        duracion         = r["Duracion"]
        hora_inicio      = r["Hora_Inicio"]
        hora_fin         = r["Hora_Fin"]
        fecha_fin        = r["Fecha_Fin"]

        horas_por_sesion = (hora_fin.seconds - hora_inicio.seconds) / 3600
        total_sesiones   = int(duracion / horas_por_sesion)

        # Solo calcular asistencia si el curso ya finalizó
        if fecha_fin < hoy:
            cumple = random.random() < 0.85

            if cumple:
                min_sesiones     = max(1, round(total_sesiones * 0.80))
                asistencia_total = random.randint(min_sesiones, total_sesiones)
            else:
                min_sesiones     = max(1, round(total_sesiones * 0.40))
                max_sesiones     = max(1, round(total_sesiones * 0.79))
                asistencia_total = random.randint(min_sesiones, max_sesiones)

            porcentaje = round((asistencia_total / total_sesiones) * 100, 2)
        else:
            asistencia_total = None
            porcentaje       = None

        cursor.execute("""
            INSERT INTO asistencia_estudiante
            (ID_Matricula, Total_Sesiones, Asistencia_Total, Porcentaje_Asistencia)
            VALUES (%s, %s, %s, %s)
        """, (id_matricula, total_sesiones, asistencia_total, porcentaje))

        insertados += 1

    conexion.commit()
    print(f"✅ {insertados} registros de asistencia insertados correctamente.")

    # Validación rápida
    cursor.execute("""
        SELECT
            SUM(CASE WHEN Porcentaje_Asistencia >= 80 THEN 1 ELSE 0 END) AS Aprobaron,
            SUM(CASE WHEN Porcentaje_Asistencia < 80  THEN 1 ELSE 0 END) AS No_Aprobaron,
            ROUND(AVG(Porcentaje_Asistencia), 2)                          AS Promedio_Asistencia
        FROM asistencia_estudiante
    """)
    v = cursor.fetchone()
    print(f"\n📊 Validación asistencia:")
    print(f"   Cumplen 80%+  : {v['Aprobaron']}")
    print(f"   No cumplen    : {v['No_Aprobaron']}")
    print(f"   Promedio      : {v['Promedio_Asistencia']}%")

except Exception as e:
    conexion.rollback()
    print("❌ Error:", e)

finally:
    cursor.close()
    conexion.close()
    print("🔒 Conexión cerrada.")
```

```python
# ======================================================
# 13. CERTIFICADOS
# ======================================================

import mysql.connector
import random
import string
from datetime import datetime, timedelta

print("🔎 Regenerando certificados correctamente (lógica estricta)...")

try:
    conexion = mysql.connector.connect(
        host="localhost",
        port=3306,
        user="root",
        password="Root8",
        database="caja_de_compensacion"
    )

    cursor = conexion.cursor(dictionary=True)
    hoy = datetime.today().date()

    # 🔥 Limpiar tabla
    cursor.execute("DELETE FROM certificados")
    conexion.commit()
    print("🗑 Tabla certificados limpiada.")

    def generar_url():
        codigo = ''.join(random.choices(string.ascii_letters + string.digits, k=25))
        return f"https://certificados.cajacompensacion.com/{codigo}"

    query = """
SELECT 
    m.ID_Matricula,
    h.Fecha_Inicio,
    h.Fecha_Fin,
    c.Costo,
    ce.Calificacion,
    IFNULL(SUM(CASE WHEN p.Estado = 'Completado' THEN p.Monto ELSE 0 END),0) AS Total_Pagado,
    ae.Porcentaje_Asistencia
FROM matricula m
JOIN curso c ON m.ID_Curso = c.ID_Curso
JOIN horario_curso h ON c.ID_Curso = h.ID_Curso
LEFT JOIN calificacion_del_estudiante ce 
    ON m.ID_Matricula = ce.ID_Matricula
LEFT JOIN pago p 
    ON m.ID_Matricula = p.ID_Matricula
LEFT JOIN asistencia_estudiante ae
    ON m.ID_Matricula = ae.ID_Matricula
GROUP BY 
    m.ID_Matricula,
    h.Fecha_Inicio,
    h.Fecha_Fin,
    c.Costo,
    ce.Calificacion,
    ae.Porcentaje_Asistencia
"""

    cursor.execute(query)
    registros = cursor.fetchall()

    for r in registros:

        id_matricula = r["ID_Matricula"]
        fecha_inicio = r["Fecha_Inicio"]
        fecha_fin = r["Fecha_Fin"]
        costo = float(r["Costo"])
        calificacion = r["Calificacion"]
        total_pagado = float(r["Total_Pagado"])
        porcentaje_asistencia = r["Porcentaje_Asistencia"]

        estado = None
        fecha_emision = None
        url = None

        # ==============================
        # 1️⃣ CURSO POR INICIAR
        # ==============================
        if fecha_inicio > hoy:
            estado = "curso por iniciar"

        # ==============================
        # 2️⃣ FORMACIÓN EN CURSO
        # ==============================
        elif fecha_fin >= hoy:
            estado = "formación en curso"

        # ==============================
        # 3️⃣ CURSO FINALIZADO
        # ==============================
        else:
            if calificacion is None or calificacion < 3.5:
                estado = "no aprobó el curso"
            elif porcentaje_asistencia is None or porcentaje_asistencia < 80:
                estado = "asistencia insuficiente"
            else:
                if total_pagado < costo:
                    estado = "pago incompleto"
                else:
                    estado = "generado"
                    fecha_emision = fecha_fin + timedelta(days=15)
                    url = generar_url()

        insert = """
        INSERT INTO certificados
        (ID_Matricula, URL_Certificado, Fecha_Emision, Estado)
        VALUES (%s,%s,%s,%s)
        """

        cursor.execute(insert, (
            id_matricula,
            url,
            fecha_emision,
            estado
        ))

    conexion.commit()
    print("✅ Certificados regenerados correctamente con validaciones estrictas.")

except Exception as e:
    print("❌ Error:", e)

finally:
    cursor.close()
    conexion.close()
    print("🔒 Conexión cerrada.")
```

```python
# ==============================
# 14. CALIFICACION_AL_CURSO
# ==============================

import mysql.connector
from datetime import datetime, timedelta
import random

try:
    conexion = mysql.connector.connect(
        host="localhost",
        port=3306,
        user="root",
        password="Root8",
        database="caja_de_compensacion"
    )

    cursor = conexion.cursor()
    hoy = datetime.today().date()

    print("Conexión exitosa")

    cursor.execute("""
        SELECT 
            m.ID_Matricula,
            MAX(h.Fecha_Fin) AS Fecha_Fin
        FROM Matricula m
        JOIN Calificacion_Del_Estudiante cde 
            ON m.ID_Matricula = cde.ID_Matricula
        JOIN Curso c ON m.ID_Curso = c.ID_Curso
        JOIN Horario_Curso h ON c.ID_Curso = h.ID_Curso
        WHERE h.Fecha_Fin < %s
        GROUP BY m.ID_Matricula
    """, (hoy,))

    matriculas_finalizadas = cursor.fetchall()
    print(f"Matrículas que pueden evaluar el curso: {len(matriculas_finalizadas)}")

    comentarios_positivos = [
        "El curso cumplió totalmente mis expectativas y los contenidos fueron claros.",
        "La metodología del formador facilitó el aprendizaje.",
        "El material de apoyo fue útil y bien estructurado.",
        "El curso estuvo bien organizado y fue fácil de seguir.",
        "Los temas abordados fueron pertinentes y aplicables."
    ]

    comentarios_neutros = [
        "El curso fue adecuado, aunque puede mejorar en algunos aspectos.",
        "Cumplió parcialmente mis expectativas.",
        "El contenido fue útil pero podría profundizar más en algunos temas.",
        "La experiencia fue aceptable, aunque con oportunidades de mejora."
    ]

    comentarios_negativos = [
        "El curso no cumplió mis expectativas.",
        "La metodología podría mejorar considerablemente.",
        "El contenido no fue lo suficientemente claro.",
        "Considero que faltó organización en el desarrollo del curso."
    ]

    registros_insertados = 0

    for id_matricula, fecha_fin in matriculas_finalizadas:

        # =========================
        # Calificación y comentario
        # =========================
        prob = random.random()

        if prob <= 0.85:
            calificacion = round(random.uniform(4.0, 5.0), 2)
            comentario   = random.choice(comentarios_positivos)
        elif prob <= 0.93:
            calificacion = round(random.uniform(3.0, 3.9), 2)
            comentario   = random.choice(comentarios_neutros)
        else:
            calificacion = round(random.uniform(1.0, 2.9), 2)
            comentario   = random.choice(comentarios_negativos)

        fecha_calificacion = fecha_fin + timedelta(days=random.randint(1, 20))

        # =========================
        # Recomendación
        # Calificación > 4.0 → siempre Sí
        # Entre 3.0 y 4.0  → 60% Sí
        # Menor a 3.0      → 15% Sí
        # =========================
        if calificacion > 4.0:
            recomendacion = "Sí"
        elif calificacion >= 3.0:
            recomendacion = random.choices(
                ["Sí", "No"],
                weights=[60, 40]
            )[0]
        else:
            recomendacion = random.choices(
                ["Sí", "No"],
                weights=[15, 85]
            )[0]

        # =========================
        # Fidelización
        # Calificación >= 4.5 → 95% Sí
        # Entre 4.0 y 4.5    → 80% Sí
        # Entre 3.0 y 4.0    → 50% Sí
        # Menor a 3.0        → 15% Sí
        # =========================
        if calificacion >= 4.5:
            fidelizacion = random.choices(
                ["Sí", "No"],
                weights=[95, 5]
            )[0]
        elif calificacion >= 4.0:
            fidelizacion = random.choices(
                ["Sí", "No"],
                weights=[80, 20]
            )[0]
        elif calificacion >= 3.0:
            fidelizacion = random.choices(
                ["Sí", "No"],
                weights=[50, 50]
            )[0]
        else:
            fidelizacion = random.choices(
                ["Sí", "No"],
                weights=[15, 85]
            )[0]

        sql = """
        INSERT INTO Calificacion_Al_Curso
        (ID_Matricula, Calificacion, Fecha_Calificacion, Comentario,
         Recomendacion, Fidelizacion)
        VALUES (%s, %s, %s, %s, %s, %s)
        """

        cursor.execute(sql, (
            id_matricula,
            calificacion,
            fecha_calificacion,
            comentario,
            recomendacion,
            fidelizacion
        ))

        registros_insertados += 1

    conexion.commit()
    print(f"✅ Evaluaciones al curso insertadas correctamente: {registros_insertados}")

except Exception as e:
    conexion.rollback()
    print("❌ Error:", e)

finally:
    cursor.close()
    conexion.close()
    print("🔒 Conexión cerrada.")
```
